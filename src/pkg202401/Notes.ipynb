Implement these methods and make any necessary adjustments to fit them into your `Array` class.

1. **Method Name:** `contains`
   - **Purpose:** Check if a specific item exists in the array.
   - **Parameters:** `int item`
   - **Return Type:** `boolean`

2. **Method Name:** `size`
   - **Purpose:** Get the current number of elements in the array.
   - **Parameters:** None
   - **Return Type:** `int`

3. **Method Name:** `isEmpty`
   - **Purpose:** Check if the array is empty.
   - **Parameters:** None
   - **Return Type:** `boolean`

4. **Method Name:** `clear`
   - **Purpose:** Clear all elements from the array.
   - **Parameters:** None
   - **Return Type:** `void`

5. **Method Name:** `get`
   - **Purpose:** Get the element at a specific index.
   - **Parameters:** `int index`
   - **Return Type:** `int`

6. **Method Name:** `set`
   - **Purpose:** Set the value of an element at a specific index.
   - **Parameters:** `int index, int value`
   - **Return Type:** `void`

7. **Method Name:** `remove`
   - **Purpose:** Remove a specific item from the array.
   - **Parameters:** `int item`
   - **Return Type:** `boolean`

8. **Method Name:** `addAll`
   - **Purpose:** Add all elements from another array to the current array.
   - **Parameters:** `Array otherArray`
   - **Return Type:** `void`

9. **Method Name:** `indexOfLast`
   - **Purpose:** Find the index of the last occurrence of a specific item in the array.
   - **Parameters:** `int item`
   - **Return Type:** `int`

10. **Method Name:** `subArray`
    - **Purpose:** Create a sub-array from the current array based on a specified range.
    - **Parameters:** `int startIndex, int endIndex`
    - **Return Type:** `Array`

```java
public class ArrayTest {
    public static void main(String[] args) {
        // Create an array with an initial capacity of 5
        Array myArray = new Array(5);

        // Insert elements into the array
        myArray.insert(10);
        myArray.insert(20);
        myArray.insert(30);
        myArray.insert(40);
        myArray.insert(50);

        // Test the contains method
        System.out.println("Contains 30: " + myArray.contains(30));  // Should print true
        System.out.println("Contains 15: " + myArray.contains(15));  // Should print false

        // Test the size method
        System.out.println("Array Size: " + myArray.size());  // Should print 5

        // Test the isEmpty method
        System.out.println("Is Array Empty: " + myArray.isEmpty());  // Should print false

        // Test the clear method
        myArray.clear();
        System.out.println("Is Array Empty (after clear): " + myArray.isEmpty());  // Should print true

        // Insert new elements after clearing
        myArray.insert(5);
        myArray.insert(10);
        myArray.insert(15);

        // Test the get method
        System.out.println("Element at index 1: " + myArray.get(1));  // Should print 10

        // Test the set method
        myArray.set(1, 20);
        System.out.println("Element at index 1 (after set): " + myArray.get(1));  // Should print 20

        // Test the remove method
        System.out.println("Remove 10: " + myArray.remove(10));  // Should print true
        System.out.println("Remove 25: " + myArray.remove(25));  // Should print false

        // Test the addAll method
        Array anotherArray = new Array(3);
        anotherArray.insert(100);
        anotherArray.insert(200);
        anotherArray.insert(300);

        myArray.addAll(anotherArray);
        System.out.println("Array Size (after addAll): " + myArray.size());  // Should print 5 + 3 = 8

        // Test the indexOfLast method
        myArray.insert(200);
        System.out.println("Index of last occurrence of 200: " + myArray.indexOfLast(200));  // Should print 8

        // Test the subArray method
        Array subArray = myArray.subArray(1, 5);
        System.out.println("Sub-Array Size: " + subArray.size());  // Should print 5 - 1 + 1 = 5
    }
}
```
